```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def collect(year):
    url = "https://www.interbrand.com/best-brands/best-global-brands/%s/ranking" % str(year)
    html = requests.get(url)
    page = BeautifulSoup(html.text, features="html5lib")
    return page
```

```python
def scrape(year, page, attributes):
    content = page.find_all("li", class_="brand-item")
    datadict = []

    for i in content:
        scraped = {}
        scraped['year'] = str(year)
        
        for attr in attributes:
            try:
                scraped[attr] = i.find("div", class_=attr).text.strip()
            except:
                scraped[attr] = ""
                
        datadict.append(scraped)
    
    scraped_data = pd.DataFrame.from_dict(datadict)    
    return scraped_data
```

```python
attributes = ("brand-rank", "brand-name", "brand-region", "brand-country", "brand-sector", "brand-value", "brand-value-change")
```

```python
df_brands = pd.DataFrame()
```

```python
for year in range(2000, 2003):
    print "collecting %s" % year  # keeping an eye on progress
    page = collect(year)
    scraped_data = scrape(year, page, attributes)
    df_brands = df_brands.append(scraped_data)
```
